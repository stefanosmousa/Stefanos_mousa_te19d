```python
#uppgift 1e
#måste importera mattematik så att jag ska kunna dra roten ur i python.
import math
punktercirkel=0
#importerade random för att jag ska kunna få fram olika punkter
import random
#måste skriva in antalet random punkter jag vill ska skrivas ut och uifrån det gör jag en "for sats" där jag har mina x och y värden för diagramet och de random nummrena skrivs ut. Efter det gjorde jag en if sats så att jag kan dra roten ur och få fram svaret
antal = 50
for i in range(antal):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)
    print(f"{x,y}")
    if math.sqrt(x**2 + y**2) < 1:
        punktercirkel += 1
#till slut får vi dividera antalet punkter som hamnar innanför cirkeln med antalet simulerade punkter och får fram ett svar på den frågan, nu är vi färdiga med denna uppgiften också
print(f"{punktercirkel/antal}")
```

```python
#1g
#först börjar man med att impotera matplotlib för att kunna rita ut diagram annars kommer python inte att förstå vad man skriver i koden och diagramet kommer inte ritas ut.
import matplotlib.pyplot as plt
#2a måste man importera matte för att man ska kunna räkna matte i python annars kommer python inte förstå vad man skriver och kommer skriva ut invalid syntax på alla täcken som har med matte att göra.
import math
#3e är att man måste importera random för att man ska kunna få fram random siffror till x och y.
import random
#efter det måste vi bestämma våra värden
punktercirkel=0
antal = 5000
#sist måste man göra en for-sats för att kunna skriva ut punkterna på diagramet, plt måste finnas inanför for satsen för att punkterna ska hamna innanför diagramet annars kommer inte punkterna ritas ut.
for i in range(antal):
    x=random.uniform(-1,1)
    y=random.uniform(-1,1)
    print(f"{x,y}")
    if math.sqrt(x**2 + y**2) < 1:
        punktercirkel += 1
        plt.plot(x,y,'*r')
    else:
        plt.plot(x,y,'*g')
```
